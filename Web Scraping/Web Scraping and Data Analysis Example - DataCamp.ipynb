```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# specify the URL containing the dataset and pass it to urlopen() to get the html of the page.
url = "http://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
```

```python
# The Beautiful Soup package is used to parse the html, that is, take the raw html text and break it into Python objects. 
# The second argument 'lxml' is the html parser whose details you do not need to worry about at this point.
soup = BeautifulSoup(html)
```

```python
# print out the first 10 rows
rows = soup.find_all('tr')
len(rows)
#for row in rows:
#    cell = row.find_all('td')
#cell
```

```python
# for each <tr> find all <td>, clean from tag format and append the pure string data into a new list
import re
list_rows=[]
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    pattern = '<.*?>' # anything between < and >
    clean = re.sub(pattern,'',str_cells) # find  tag and replace with ''
    list_rows.append(clean)
```

```python
print(len(list_rows));
list_rows[len(list_rows)-1] #last one
```

```python
# convert the list into a dataframe and get a quick view of the first 10 rows using Pandas.
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
############ Data Manipulation and Cleaning ########################################
# To clean it up, you should split the "0" column into multiple columns at the comma position.
df1 = df[0].str.split(',', expand=True)
df1.head()
```

```python
# You can use the strip() method to remove the opening square bracket on column "0."
df1[0] = df1[0].str.strip('[')
df1.head()
```

```python
# Again with the second column
df1[0] = df1[0].str.strip(']')
df1.head()
```

```python
# find table headers
col_labels = soup.find_all('th')
col_labels
```

```python
# extract text in between html tags for table headers.
all_header = []
col_str = str(col_labels)
cleanHeader = BeautifulSoup(col_str).get_text()
all_header.append(cleanHeader)
print(all_header)
```

```python
# create a dataframe of headers
df_header = pd.DataFrame(all_header)
df_header.head()
```

```python
# split the column
df_header = df_header[0].str.split(',', expand = True)
df_header.head()
```

```python
# The two dataframes can be concatenated into one using the concat() method
df_whole = pd.concat([df_header, df1])
df_whole.head()
```

```python
df_whole = df_whole.rename(columns=df_whole.iloc[0])
df_whole.head()
```

```python
# Analysing the Data
df_whole.info()
df_whole.shape
```

```python
# drop all rows with any missing values.
df_whole = df_whole.dropna(axis=0, how='any')
df_whole.head()
```

```python
# drop the duplicated first row
df_clean = df_whole.drop(df_whole.index[0])
df_clean.head()
```

```python
# rename first and last column headers
df_clean.rename(columns={'[Place':'Place'}, inplace=True)
df_clean.rename(columns={' Team]':'Team'}, inplace=True)
df_clean.head()
```

```python
#  removing the closing bracket for cells in the "Team" column.
df_clean['Team'] = df_clean['Team'].str.strip(']')
df_clean.head()
```

```python
########### Data Analysis and Visualization#################################
# what was the average finish time (in minutes) for the runners?
```

```python
# You need to convert the column "Chip Time" into just minutes. 
time_list = df_clean[' Chip Time'].tolist()
time_list
```

```python
time_mins=[]
for i in time_list:
    if len(i) == 8:
        h, m, s = i.split(':')
        math = (int(h)*3600 + int(m) * 60 + int(s)) / 60
        time_mins.append(math)
    else:
        m, s = i.split(':')
        math = (int(m) * 60 + int(s)) / 60
        time_mins.append(math)
print(time_mins)
 
```

```python
# Add a new column of time in minutes
df_clean['Runners_mins'] = time_mins
df_clean.head()
```

```python
df_clean.describe(include=[np.number]) # statistics for number columns only
```

```python
df_clean.boxplot(column='Runners_mins')
plt.grid(True, axis='y')
plt.ylabel('Chip Time')
plt.xticks([1],['Runners'])
plt.show()
```
