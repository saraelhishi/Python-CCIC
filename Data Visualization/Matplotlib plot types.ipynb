```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Bar plot
plt.bar([1,3,6,10,25], height=[4,3,6,2,9],width=.8)
```

```python
### Stacked Bar plot
N = 5
A = (20, 35, 30, 35, 27)
B = (25, 32, 34, 20, 25)

ind = np.arange(N)    # the x locations for the groups
width = 0.35       

p1 = plt.bar(ind, A, width)
p2 = plt.bar(ind, B, width, bottom=A)

plt.ylabel('Scores')
plt.title('Scores by group')
plt.xticks(ind,('T1','T2','T3','T4','T5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Group A', 'Group B'))

plt.show()
```

```python
### Grouped bar plot
fig, ax = plt.subplots()

p1 = ax.bar(ind-width/2, A, width, color='green')
p2 = ax.bar(ind+width/2, B, width, color='orange')

plt.ylabel('Scores')
plt.title('Scores by group')
plt.xticks(ind,('T1','T2','T3','T4','T5'))
plt.yticks(np.arange(0, 35, 10))
plt.legend((p1[0], p2[0]), ('Group A', 'Group B'))

plt.show()
```

```python
### Horizontal Bar plot
players=('Sam','Tom','Mark')
y_ind = np.arange(len(players))
players_scores =  10 * np.random.rand(len(players))

plt.barh(y_ind, players_scores)

plt.show()
```

```python
### Histogram plt.hist()
data = np.random.randn(100) # 100 normally distributed values
#plt.hist(data)
plt.hist(data, bins=20, alpha=0.8, histtype='bar', color='lightyellow',edgecolor='green')
plt.show()
```

```python
# comparing histograms of several distributions tips: use stepfilled style along with low transperancy value
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.3, normed=True, bins=40) # batch parameters
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
### Pie charts: the slices will be ordered and plotted counter-clockwise
labels = 'Fiber','Fat', 'carbohydrates', 'Protein'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
wedges_label = '%1.1f%%' # for string format use 'fmt%pct'
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels,
       autopct=wedges_label, shadow=True, startangle=90) 
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
### Scatter ---- 
# scatter(x, y, s=Marker size, c=color)


### Scatter plot with a range of colors values, sizes, and tranceparency in order to illustrate multidimensional data
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
marker_sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=marker_sizes, alpha=0.3)
plt.colorbar(); # show color scale
```
