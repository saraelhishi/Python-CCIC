```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

from scipy import misc
```

```python
f = misc.face() # a popular image for analysis

#misc.imsave('face.png', f) # save the image in your working directory
plt.imshow(f) # show it on python
plt.axis('off')
```

```python
f1=misc.face(gray=True)
plt.imshow(f1,cmap=plt.cm.gray)
plt.contour(f1,[50,200])
```

```python
# Image Manipulation
# look at images as arrays and use numpy to handle them
face=misc.face(gray=True)
#face[0,40]
#face[10:13,20:23]

# create a mask
face[100:120]=255
lx,ly=face.shape
X,Y=np.ogrid[0:lx,0:ly]
mask=(X-lx/2)**2+(Y-ly/2)**2>lx*ly/4
face[mask]=0
face[range(400),range(400)]=255
plt.imshow(face)
```

```python
# Geometrical Transformations
# flipping
face=misc.face(gray=True)
lx,ly=face.shape
crop=face[lx//4:-lx//4,ly//4:-ly//4]
flip=np.flipud(face)
plt.imshow(flip)
```

```python
# rotation
from scipy import ndimage
rotate=ndimage.rotate(face,45)
plt.imshow(rotate)

rotate_noreshape=ndimage.rotate(face,45,reshape=False)
plt.imshow(rotate_noreshape)
```

```python
# Filtering
# Blurring/Smoothing
face=misc.face(gray=True)
blurred=ndimage.gaussian_filter(face,sigma=3)
plt.imshow(blurred)
way_blurred=ndimage.gaussian_filter(face,sigma=5)
plt.imshow(way_blurred)
local_mean=ndimage.uniform_filter(face,size=11)
plt.imshow(local_mean)
```

```python
# Sharpening
face=misc.face(gray=True).astype(float)
blurred=ndimage.gaussian_filter(face,3)
filter_blurred=ndimage.gaussian_filter(blurred,1)
alpha=30
sharpened=blurred+alpha*(blurred-filter_blurred)
plt.imshow(sharpened)
```

```python
# Extracting Features
# Edge Detection
im=np.zeros((256,256))
im[64:-64,64:-64]=1
im=ndimage.rotate(im,15,mode='constant')
im=ndimage.gaussian_filter(im,8)
plt.imshow(im)
```

```python
sx=ndimage.sobel(im,axis=0,mode='constant')
sy=ndimage.sobel(im,axis=1,mode='constant')
sob=np.hypot(sx,sy)
plt.imshow(sob)
```

```python
# Segmentation
n,l=10,256
im=np.zeros((l,l))
np.random.seed(1)
points=l*np.random.random((2,n**2))
im[(points[0]).astype(np.int),(points[1]).astype(np.int)]=1
im=ndimage.gaussian_filter(im,sigma=l/(4.*n))
plt.imshow(im)
```

```python
mask=(im>im.mean()).astype(np.float)
mask+=0.1*im
img=mask+0.2*np.random.randn(*mask.shape)
plt.imshow(img)
```

```python
binary_img=img>0.5
plt.imshow(binary_img)
```
